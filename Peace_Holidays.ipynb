```python

# Function to check if a country is valid using pycountry
def check_country(country):
    try:
        # to look up the country in the list of countries in pycountry
        pycountry.countries.lookup(country)
        return True
    except LookupError:
        return False

# verification to each value in the "Country" column
df['Valid_Country'] = df['Country'].apply(check_country)

# Display the DataFrame with the new column indicating if the country is valid
print ( df[["Country", "Valid_Country"]])

invalid_countries = df[df['Valid_Country'] == False]

# Select and print specific columns
print(invalid_countries[["Country" ,'Valid_Country']])

# Drop rows where "Valid_Country" is False
df = df.loc[df['Valid_Country'] == True]

# Display the DataFrame with valid countries only
print(df[["Country"]])

```

```python
def check_state(country, state):
    try:
        # Look up the country
        country_obj = pycountry.countries.lookup(country)
        # Look up the state/province within that country
        subdivision = pycountry.subdivisions.get(code=f'{country_obj.alpha_2}-{state}')
        return True
    except LookupError:
        return False

# Apply the verification to each value in the DataFrame
df['Valid_State'] = df.apply(lambda row: check_state(row['Country'], row['State']), axis=1)

# Print the DataFrame with the new column indicating if the state is valid
invalid_cities = df[df['Valid_State'] == False]
print(invalid_cities[["State", "Valid_State"]])
```

```python
import seaborn as sns

top_10_activities = activity_counts.head(10).index

top_activities_df = florida_df[florida_df['Activity'].isin(top_10_activities)]
sns.countplot(data=top_activities_df, y='Activity', order=top_10_activities)
```

```python
location_attack_counts_florida = florida_df['Location'].value_counts()
top_10_locations = location_attack_counts_florida.head(10).index.tolist()
top_locations_df = florida_df[florida_df['Location'].isin(top_10_locations)]
sns.histplot(data=top_locations_df, y='Location', stat='count', bins=len(top_10_locations))
```
